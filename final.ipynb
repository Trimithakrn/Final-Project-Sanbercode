```python
# Mendefinisikan batas atas dan batas bawah menggunakan metode IQR untuk setiap kolom numerik
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Menampilkan nilai outlier untuk setiap kolom
outliers = ((data < lower_bound) | (data > upper_bound)).sum()
print("Jumlah outlier pada setiap kolom:")
print(outliers)
```

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
data = data[~((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
# Melakukan handling outlier menggunakan boxplot
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
data = data[~((data < (Q1 - 1.5 * IQR)) |
              (data > (Q3 + 1.5 * IQR))).any(axis=1)]

# Visualisasi boxplot setelah handling outlier
plt.figure(figsize=(10, 6))
data.boxplot(color=dict(boxes='b', whiskers='b', medians='b', caps='b'))
plt.title('Boxplot setelah Handling Outlier')
plt.grid(False)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data.iloc[:,1:])
pd.DataFrame(data_scaled)
```

```python
from sklearn.cluster import KMeans
# Menentukan jumlah cluster dengan metode Elbow
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(data_scaled)
    wcss.append(kmeans.inertia_)
```

```python
# Menampilkan nilai dari setiap cluster
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(data_scaled)
    clusters = kmeans.labels_
    print(f'Cluster counts for {i} clusters:',
          pd.Series(clusters).value_counts().to_dict())
```

```python
# Melakukan klastering dengan jumlah cluster tertentu
kmeans = KMeans(n_clusters=2, init='k-means++', random_state=42)
clusters = kmeans.fit_predict(data_scaled)
```

```python
data['Cluster'] = clusters
data.to_csv('hasil_cluster.csv')
```

```python
import matplotlib.pyplot as plt
# Visualisasi Elbow Method
plt.plot(range(1, 11), wcss)
plt.title('Metode Elbow')
plt.xlabel('Jumlah Clusters')
plt.ylabel('WCSS')  # Within cluster sum of squares
plt.show()
```

```python
plt.scatter(data[data['Cluster'] == 0]['Pendapatan'],
            data[data['Cluster'] == 0]['Kesehatan'], s=100, c='red', label='Cluster 1')
plt.scatter(data[data['Cluster'] == 1]['Pendapatan'],
            data[data['Cluster'] == 1]['Kesehatan'], s=100, c='blue', label='Cluster 2')

# Menandai pusat klaster
plt.title('Hasil Klastering')
plt.xlabel('Pendapatan')
plt.ylabel('Kesehatan')
plt.legend()
plt.show()
```

```python
plt.scatter(data[data['Cluster'] == 0]['Kematian_anak'],
            data[data['Cluster'] == 0]['Harapan_hidup'], s=100, c='red', label='Cluster 1')
plt.scatter(data[data['Cluster'] == 1]['Kematian_anak'],
            data[data['Cluster'] == 1]['Harapan_hidup'], s=100, c='blue', label='Cluster 2')

# Menandai pusat klaster
plt.title('Keshatan')
plt.xlabel('Kematian Anak')
plt.ylabel('Harapan Hidup')
plt.legend()
plt.show()
```
